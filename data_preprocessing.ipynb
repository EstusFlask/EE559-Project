Drop the 'veil-type' column as it has zero variance


```python
import pandas as pd
df = pd.read_csv('raw_data/mushrooms.csv')
df = df.drop('veil-type', axis=1)
```

#### Label Encoding
Encode target variable: 'e' -> 0, 'p' -> 1


```python
from sklearn.preprocessing import (LabelEncoder)

le = LabelEncoder()
df['class'] = le.fit_transform(df['class'])
```

#### Feature Encoding
- One-Hot Encoding : Since there are more than two types
- Dummy Variable Trap

```python
import pandas as pd

# Separate features (X) and target (y)
X = df.drop('class', axis=1)
y = df['class']

# Apply One-Hot Encoding to all feature columns
# drop_first=True helps avoid multicollinearity for Logistic Regression
X_encoded = pd.get_dummies(X, drop_first=True)
```

#### Train-Test Split
Split the data into 80% training and 20% testing


```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

#### Save the processed data
In order to make the results replicable we saved the results as splitted datasets

```python
X_train.to_csv("preprocessed_data/X_train.csv", index=False)
X_test.to_csv("preprocessed_data/X_test.csv", index=False)
y_train.to_csv("preprocessed_data/y_train.csv", index=False)
y_test.to_csv("preprocessed_data/y_test.csv", index=False)
```
